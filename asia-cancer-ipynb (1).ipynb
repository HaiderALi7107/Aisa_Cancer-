```python
# Cancer Dataset Generator and Visualizer for Asia (Jupyter Notebook Format)

# Cell 1: Imports and Configuration
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)  # For reproducibility
```

```python
# Cell 2: Define Constants
countries = [
    "India", "China", "Japan", "Indonesia", "Pakistan",
    "Bangladesh", "Philippines", "Vietnam", "Thailand", "Malaysia"
]
cancer_types = ["Lung", "Breast", "Colorectal", "Stomach", "Liver"]
n = 100  # Number of patients
```

```python
# Cell 3: Generate Dataset
patients = {
    "ID": [f"P{i+1:04d}" for i in range(n)],
    "Country": np.random.choice(countries, size=n),
    "Age": np.random.randint(25, 85, size=n),
    "Gender": np.random.choice(["Male", "Female"], size=n),
    "Cancer": np.random.choice(cancer_types, size=n),
    "Stage": np.random.choice(["I", "II", "III", "IV"], size=n, p=[0.2, 0.3, 0.3, 0.2]),
    "Treatment": np.random.choice(["Surgery", "Chemotherapy", "Radiation", "Immunotherapy"], size=n),
    "Survival_Years": np.clip(np.round(np.random.exponential(scale=3, size=n), 1), 0, 15),
    "Smoker": np.random.choice(["Yes", "No"], size=n),
    "FamilyHistory": np.random.choice(["Yes", "No"], size=n)
}
df = pd.DataFrame(patients)
```

```python
# Cell 4: Save Dataset
df.to_csv("cancer_asia_2025.csv", index=False)
df.head()  # Preview first few rows
```

```python
# Cell 5: Visualization Setup
sns.set(style="whitegrid")
```

```python
# Cell 6: Cancer Type Distribution
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x="Cancer", order=df["Cancer"].value_counts().index)
plt.title("Cancer Type Distribution in Asia")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cell 7: Age Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df["Age"], bins=10, kde=True)
plt.title("Age Distribution")
plt.tight_layout()
plt.show()
```

```python
# Cell 8: Survival Years by Stage
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x="Stage", y="Survival_Years")
plt.title("Survival Years by Stage")
plt.tight_layout()
plt.show()
```

```python
# Cell 9: Patient Count by Country
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x="Country", order=df["Country"].value_counts().index)
plt.title("Patient Count per Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cell 10: Correlation Heatmap
plt.figure(figsize=(5, 3))
sns.heatmap(df[["Age", "Survival_Years"]].corr(), annot=True, cmap="coolwarm")
plt.title("Age vs. Survival Correlation")
plt.tight_layout()
plt.show()
```
